# Equilibration of ligands in free stage

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sma(data, n, window):
    """
    Function to calculate the simple moving average. 
    @param data: array of data to aberage
    @param n: number of datapoints
    @param window: time window over which to calculate SMA
    @return running average
    """
    s_i = 0
    cumulative_sum = []
    for i in range(n):
        s_i = s_i + data[i]
        cumulative_sum.append(s_i)
    cumulative_sum = np.array(cumulative_sum)    
    running_avg = (cumulative_sum[window:] - cumulative_sum[:-window]) / float(window)
    return running_avg
```

## Minimisation

```python
ligands = np.arange(1, 17, 1)
for ligand in ligands:
    path = f"equilibration/ligands/lig_h_{ligand}/"
    time,energy = np.loadtxt(f"{path}/potential.xvg",comments=["#","@"],unpack=True)
    plt.scatter(time, energy, marker=".")
    plt.legend(f"{ligand}")
    plt.show()
```

## NVT

```python
ligands = np.arange(1, 17, 1)
for ligand in ligands:
    print(ligand)
    path = f"equilibration/ligands/lig_h_{ligand}/"
    try:
        time,temperature = np.loadtxt(f"{path}/temperature.xvg",comments=["#","@"],unpack=True)
        plt.plot(time, temperature, c="black")
        plt.plot(time[50:], sma(temperature, len(temperature), 50), c="red")
        plt.show()
    except IndexError:
        print(f"error with temperature file on ligand {ligand}")
```

## NPT Pressure

```python
ligands = np.arange(1, 17, 1)
for ligand in ligands:
    print(ligand)
    path = f"equilibration/ligands/lig_h_{ligand}/"
    try:
        time,pressure = np.loadtxt(f"{path}/pressure.xvg",comments=["#","@"],unpack=True)
        plt.plot(time, pressure, c="black")
        plt.plot(time[50:], sma(pressure, len(pressure), 50), c="red")
        plt.show()
    except FileNotFoundError:
        print(f"pressure file for ligand {ligand} not found")
```

## NPT Density

```python
ligands = np.arange(1, 17, 1)
for ligand in ligands:
    print(ligand)
    path = f"equilibration/ligands/lig_h_{ligand}/"
    try:
        time,density = np.loadtxt(f"{path}/density.xvg",comments=["#","@"],unpack=True)
        plt.plot(time, density, c="black")
        plt.plot(time[50:], sma(density, len(density), 50), c="red")
        plt.show()
    except FileNotFoundError:
        print(f"density file for ligand {ligand} not found")
```
